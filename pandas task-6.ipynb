### column : Age
* detecting and treating missing values
* typecasting
* detecting and replacing outliers
* variance
* scaling

```python
df[df['Age'].isnull()]
```

```python
df['Age'].fillna(method = 'ffill',inplace = True)
```

```python
df['Age'] = df['Age'].astype('int')
```

```python
df.info()
```

```python
df['Age'].median()
```

```python
df['Age'] = na.where((df['Age']<0)|(df['Age']>100),44,df['Age'])
```

```python
df['Age'].mean()
```

```python
df['Age'].median()
```

```python
# as both are close values so it is normalization
def normal(n):
    a = (n - df['Age'].min())/(df['Age'].max()-df['Age'].min())
    return a
```

```python
df['Age'].apply(normal)
```
